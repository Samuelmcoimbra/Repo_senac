```python
import numpy as np 
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df = pd.DataFrame({
    'Ano': [2020, 2021, 2022, 2023],
    'Vendas': [100, 150, 130, 180]})

df['Ano'] = df['Ano'].astype(str) # Correção de tipo de caracter de int para str
# A mudança só pode ser feita com variáveis qualitativas
```

```python
plt.plot(df['Ano'], df['Vendas'])
plt.title('Vendas por Ano')
plt.xlabel('Ano')
plt.ylabel('Vendas')
plt.grid(True)
plt.show()
```

```python
df = pd.DataFrame({
    'Produto': ['A', 'B', 'C'],
    'Vendas': [120, 90, 150]})
df.sort_values(by= 'Vendas', ascending= False, inplace= True)
plt.bar(df['Produto'], df['Vendas'])
plt.title('Vendas  por Produto')
plt.xlabel('Produto')
plt.ylabel('Vendas')
plt.show
```

```python
dados = {
    'Ano': [2018, 2019, 2020, 2021, 2022],
    'Vendas': [150, 200, 250, 300, 350]}
```

```python
df_dias = pd.DataFrame({'Dias': [ 'Segunda', 'Terça', 'Quarta', 'Quinta', 'Sexta'],
      'Venda': [500, 200, 1000, 150, 1950]})

df_produtos = pd.DataFrame({'Produtos': ['A', 'B', 'C'],
                            'Vendas': [120, 90, 150]})

# Gráfico 1 (0,0)
fig, axs = plt.subplots(2,2, figsize= (12,8))
axs[0, 0].plot(df_dias['Dias'], df_dias['Venda'], linestyle= '-', color= 'crimson')
axs[0,0].set_title('Vendas por Dias')
axs[0,0].set_xlabel('Dia da Semana')
axs[0,0].set_ylabel('Valor Vendido')

# Gráfico 2 (0,1)

axs[0,1].bar(df_produtos['Produtos'], df_produtos['Vendas'],   linestyle= '-', color= 'green')
axs[0,1].set_title('Vendas de Produtos')
axs[0,1].set_xlabel('Produtos')
axs[0,1].set_ylabel('Vendas')

# Gráfico 3 (1,0)
axs[1,0].barh(df_produtos['Produtos'], df_produtos['Vendas'],  linestyle= '-', color= 'black')
axs[1,0].set_title('Vendas de Produtos (Horizontal)')
axs[1,0].set_xlabel('Vendas')
axs[1,0].set_ylabel('Produtos')

# Gráfico 4 (1,1)
axs[1,1].axis('off')
axs[1,1].text(0.1,1, 'Resumo da Aula:', fontsize=14, color= 'black') # posicionamento no .text(x, y, texto)
axs[1,1].text(0.1, 0.6, '- subplots usa matriz de eixos', fontsize=12, color= 'black')
axs[1,1].text(0.1, 0.5, '- mais controle sobre cada gráfico', fontsize=12, color= 'black')
axs[1,1].text(0.1, 0.4, '- ideal para projetos mais complexos', fontsize=12, color= 'black')
plt.tight_layout()
```
