```python
import numpy as np 
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df = pd.DataFrame({
    'Ano': [2020, 2021, 2022, 2023],
    'Vendas': [100, 150, 130, 180]})

df['Ano'] = df['Ano'].astype(str) # Correção de tipo de caracter de int para str
# A mudança só pode ser feita com variáveis qualitativas
```

```python
plt.plot(df['Ano'], df['Vendas'])
plt.title('Vendas por Ano')
plt.xlabel('Ano')
plt.ylabel('Vendas')
plt.grid(True)
plt.show()
```

```python
df = pd.DataFrame({
    'Produto': ['A', 'B', 'C'],
    'Vendas': [120, 90, 150]})
df.sort_values(by= 'Vendas', ascending= False, inplace= True)
plt.bar(df['Produto'], df['Vendas'])
plt.title('Vendas  por Produto')
plt.xlabel('Produto')
plt.ylabel('Vendas')
plt.show
```

```python
dados = {
    'Ano': [2018, 2019, 2020, 2021, 2022],
    'Vendas': [150, 200, 250, 300, 350]}
```

```python
df = pd.DataFrame(dados)
```

```python
plt.figure(figsize=(8,5))
plt.barh(df['Ano'], df ['Vendas'], color= 'limegreen')
plt.title('Vendas por Ano')
plt.xlabel('Vendas')
plt.ylabel('Ano')
plt.savefig('grafico_arras_horizontais.png')
plt.show()
```

```python
df = {'dias': [ 'Segunda', 'Terça', 'Quarta', 'Quinta', 'Sexta'],
      'Venda': [500, 200, 1000, 150, 1950]}

dias =  df['dias']
valores = df['Venda']
```

```python
plt.plot(dias, valores, marker= 'o', linestyle= '-', color= 'crimson') # marker = marcador, linestyle = a linha do gráfico
plt.title('Vendas por Dias')
plt.xlabel('Dia da Semana')
plt.ylabel('Valor vendido')

plt.show()
```
