```python
df = pd.read_csv('https://www.gov.br/anp/pt-br/centrais-de-conteudo/dados-abertos/arquivos/shpc/dsas/ca/ca-2025-01.zip', sep= ';', encoding= 'latin1')
df
```

```python
df_rj = df.loc[df['Estado - Sigla'] == 'RJ']
df_rj
```

```python
df_rj['Valor de Venda'] = df_rj['Valor de Venda'].str.replace('R$', '').str.replace('.', '').str.replace(',', '.').astype(float)



```

```python
df_cidade_rj = df_rj.loc[df_rj['Municipio'] == 'RIO DE JANEIRO']
df_cidade_rj
```

```python
estudo_gas = df_cidade_rj.loc[df_cidade_rj['Produto'] == 'GASOLINA']
estudo_gas
```

```python
df_gas_tab_rj = estudo_gas.groupby('Bairro')['Valor de Venda'] .mean().reset_index()
df_gas_tab_rj = df_gas_tab_rj.sort_values(by='Valor de Venda', ascending=False)
df_gas_tab_rj
```

```python
gas_array = np.array(df_gas_tab_rj ['Valor de Venda'])
q1 = np.percentile(gas_array , 25)
q2 = np.percentile(gas_array , 50)
q3 = np.percentile(gas_array , 75)
media = np.mean(gas_array)
mediana = np.median(gas_array)

distancia =  (media - mediana) / mediana
iqr = q3 - q1
limite_superior = q3 + (1.5 * iqr)
```

```python
print(f'A distância é {distancia*100:.2f}')
print(f'O iqr é {iqr}')
print(f'O limite superior é {limite_superior:.2f}')
```

```python
gas_rj_outliers = df_gas_tab_rj.loc[df_gas_tab_rj['Valor de Venda'] >=limite_superior]
gas_rj_outliers.sort_values(by= 'Valor de Venda', ascending= False)
```
