```python
estudo = pd.read_csv('Preços semestrais.csv', sep= ';', encoding= 'latin1')
estudo
```

```python
estudo_rj = estudo.loc[estudo['Estado - Sigla'] == 'RJ']
estudo_rj
```

```python
estudo_rj['Valor de Venda'] = estudo_rj['Valor de Venda'].str.replace('R$', '').str.replace('.', '').str.replace(',', '.').astype(float)



```

```python
estudo_cidade_rj = estudo_rj.loc[estudo_rj['Municipio'] == 'RIO DE JANEIRO']
estudo_cidade_rj
```

```python
estudo_gas = estudo_cidade_rj.loc[estudo_cidade_rj['Produto'] == 'GASOLINA']
estudo_gas
```

```python
estudo_gas_tab = estudo_gas.groupby('Bairro')['Valor de Venda'] .mean().reset_index()
estudo_gas_tab = estudo_gas_tab.sort_values(by='Valor de Venda', ascending=False)
estudo_gas_tab
```

```python
estudo_gas_tab.head()
```

```python
estudo_gas_tab.tail()
```
