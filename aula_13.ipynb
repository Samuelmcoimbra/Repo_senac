## Estatística


```python
# Quartil é uma linha de corte e não (Necessariamente) um valor presente na tabela
# Quartis dividem os dados em quatro partes Q1, Q2 e Q3
# Q1 = 25%
# Q2 = 50% (Q2 é o mesmo valor da mediana)
# Q3 = 75% do conjunto
# Biblioteca numpy (import numpy as np)
```

```python
import numpy as np # np = numpy
import pandas as pd
```

```python
dados = np.array([12, 15, 17, 20, 22, 25, 28, 30, 35, 40])
q1 = np.percentile(dados, 25)
q2 = np.percentile(dados, 50)
q3 = np.percentile(dados, 75)
mediana = np.median(dados) # median = mediana

print(f'Primeiro quartil (Q1): {q1}')
print(f'Primeiro quartil (Q2): {q2}')
print(f'Primeiro quartil (Q3): {q3}')
print(f'Mediana {mediana}')
```
