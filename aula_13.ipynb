```python
# Quartil é uma linha de corte e não (Necessariamente) um valor presente na tabela
# Quartis dividem os dados em quatro partes Q1, Q2 e Q3
# Q1 = 25%
# Q2 = 50% (Q2 é o mesmo valor da mediana)
# Q3 = 75% do conjunto
# Biblioteca numpy (import numpy as np)
```

```python
import numpy as np # np = numpy
import pandas as pd
import matplotlib.pyplot as plt
```

```python
dados = np.array([12, 15, 17, 20, 22, 25, 28, 30, 35, 40])
q1 = np.percentile(dados, 25)
q2 = np.percentile(dados, 50)
q3 = np.percentile(dados, 75)
mediana = np.median(dados) # median = mediana

print(f'Primeiro quartil (Q1): {q1}')
print(f'Primeiro quartil (Q2): {q2}')
print(f'Primeiro quartil (Q3): {q3}')
print(f'Mediana {mediana}')
```

```python
região_metropolitana = ['Rio de Janeiro', 'Belford Roxo',' Cachoeira de Macacu', 'Duque de Caxias', 'Guapimirim', 'Itaboraí', 'Itaguaí', 'Japeri', 'Magé', 
    'Maricá', 'Mesquita', 'Nilópolis', 'Niterói', 'Nova Iguaçu', 'Paracambi', 'Petrópolis', 'Queimados', 'Rio Bonito', 'Seropédica',    
    'São Gonçalo', 'São João de Meriti', 'Tanguá']
```

```python
região_metropolitana_decada = dado.loc[(dado['munic'].isin(região_metropolitana)) & (dado['ano'] >= 2015)]
```

```python
veiculos_capital = região_metropolitana_decada.groupby('munic')['furto_veiculos'].sum().reset_index()
veiculos_capital.sort_values(by= 'furto_veiculos', ascending= False).reset_index()
```

```python
veiculos_array = np.array(veiculos_capital ['furto_veiculos'])
q1 = np.percentile(veiculos_array , 25)
q2 = np.percentile(veiculos_array , 50)
q3 = np.percentile(veiculos_array , 75)
media = np.mean(veiculos_array)
mediana = np.median(veiculos_array)
```

```python
distancia =  (media - mediana) / mediana
iqr = q3 - q1
limite_superior = q3 + (1.5 * iqr)
```

```python
veiculos_outliers = veiculos_capital.loc[veiculos_capital['furto_veiculos'] >=limite_superior]
veiculos_outliers.sort_values(by= 'furto_veiculos', ascending= False)
```

```python
veiculos_capital_rj = veiculos_capital.sort_values(by= 'furto_veiculos', ascending= False).reset_index().head()
```

```python
plt.figure(figsize=(8,5))
plt.bar(veiculos_capital_rj['munic'], veiculos_capital_rj['furto_veiculos'])
plt.title('furtos por municipio')
plt.xlabel('munic')
plt.ylabel('furto_veiculos')
plt.show()
```

```python
gestao1 = região_metropolitana_gest1.groupby('cisp')['estelionato'].sum().reset_index()
gestao1.sort_values(by= 'estelionato', ascending= False).head()
```

```python
gestao1_array = np.array(gestao1 ['estelionato'])
q1 = np.percentile(gestao1_array , 25)
q2 = np.percentile(gestao1_array , 50)
q3 = np.percentile(gestao1_array , 75)
media = np.mean(gestao1_array)
mediana = np.median(gestao1_array)
distancia =  (media - mediana) / mediana
iqr = q3 - q1
limite_superior = q3 + (1.5 * iqr)
```

```python
gestao1_outliers = gestao1.loc[gestao1['estelionato'] >=limite_superior]
gestao1_outliers.sort_values(by= 'estelionato', ascending= False)
```

```python
plt.figure(figsize=(8,5))
plt.bar(gestao1['cisp'], gestao1['estelionato'])
plt.title('Estelionato por munic')
plt.xlabel('munic')
plt.ylabel('estelionato')
plt.show()
```
