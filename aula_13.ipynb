## Estatística


```python
# Quartil é uma linha de corte e não (Necessariamente) um valor presente na tabela
# Quartis dividem os dados em quatro partes Q1, Q2 e Q3
# Q1 = 25%
# Q2 = 50% (Q2 é o mesmo valor da mediana)
# Q3 = 75% do conjunto
# Biblioteca numpy (import numpy as np)
```

```python
import numpy as np # np = numpy
import pandas as pd
```

```python
dados = np.array([12, 15, 17, 20, 22, 25, 28, 30, 35, 40])
q1 = np.percentile(dados, 25)
q2 = np.percentile(dados, 50)
q3 = np.percentile(dados, 75)
mediana = np.median(dados) # median = mediana

print(f'Primeiro quartil (Q1): {q1}')
print(f'Primeiro quartil (Q2): {q2}')
print(f'Primeiro quartil (Q3): {q3}')
print(f'Mediana {mediana}')
```

```python
df = pd.read_csv('pesquisa_satisfacao.csv', sep= ';')
df
```

```python
tempo_espera_array = np.array(df['Tempo de Espera (min)'])
```

```python
q1 = np.percentile(tempo_espera_array , 25)
q2 = np.percentile(tempo_espera_array , 50)
q3 = np.percentile(tempo_espera_array , 75)
media = np.mean(tempo_espera_array)
mediana = np.median(tempo_espera_array)
```

```python
print(f'Primeiro quartil (Q1): {q1}')
print(f'Primeiro quartil (Q2): {q2}')
print(f'Primeiro quartil (Q3): {q3}')
print(f'Mediana {mediana}')
print(media)
```

```python
dado = pd.read_csv('https://www.ispdados.rj.gov.br/Arquivos/BaseDPEvolucaoMensalCisp.csv', sep= ';', encoding= 'latin1')
dado
```

```python
delegacias = dado.groupby('cisp')['furto_celular'].sum().reset_index()
delegacias_tab = delegacias.sort_values(by='furto_celular', ascending=False).reset_index()
delegacias_tab
```

```python
delegacias_array = np.array(delegacias_tab['furto_celular'])

```

```python
q1 = np.percentile(delegacias_array , 25)
q2 = np.percentile(delegacias_array , 50)
q3 = np.percentile(delegacias_array , 75)
media = np.mean(delegacias_array)
mediana = np.median(delegacias_array)
```

```python
print(f'Primeiro quartil (Q1): {q1}')
print(f'Primeiro quartil (Q2): {q2}')
print(f'Primeiro quartil (Q3): {q3}')
print(f'Mediana {mediana}')
print(media)
```
