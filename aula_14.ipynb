## MEDIDAS DE TENDÊNCIA CENTRAL

```python
# distancia =  (media - mediana) / mediana
# distancia = menor de 10% a média é confiavel 
# distancia = > 10% < 25% a média é moderadamente influciada por valores extremos
# distancia = > 25% a média é muito alterada por valores extremos
```

```python
import numpy as np 
import pandas as pd
```

```python
df = pd.read_csv('funcionarios_dados_ti.csv', sep= ';')
df
```

```python
df_array = np.array(df['Salario'])
```

```python
media = np.mean(df_array)
mediana = np.median(df_array)
```

```python
print(f'A média é {media}')
print(f'A mediana é {mediana}')
```

```python
distancia =  (media - mediana) / mediana
print(f'a distância é de {round(distancia*100,2)}%')
```
